## Data Processing

```python
# Read Data
WETNWILD2019 = pd.read_csv("WETNWILD2019.csv", sep = "\t", usecols = ["language","tweet"])
WETNWILD2020 = pd.read_csv("WETNWILD2020.csv", sep = "\t", usecols = ["language","tweet"])
WETNWILD2021 = pd.read_csv("WETNWILD2021.csv", sep = "\t", usecols = ["language","tweet"])
```

```python
# Keep only English rows
WETNWILD2019 = WETNWILD2019[WETNWILD2019["language"]=="en"]
WETNWILD2019.index=range(len(WETNWILD2019))
WETNWILD2019=WETNWILD2019['tweet']

WETNWILD2020 = WETNWILD2020[WETNWILD2020["language"]=="en"]
WETNWILD2020.index=range(len(WETNWILD2020))
WETNWILD2020=WETNWILD2020['tweet']

WETNWILD2021 = WETNWILD2021[WETNWILD2021["language"]=="en"]
WETNWILD2021.index=range(len(WETNWILD2021))
WETNWILD2021=WETNWILD2021['tweet']
```

```python
def textPreprocessing(text) :
    
    # to lower case
    text = text.lower()
    
    # delete url
    text = re.sub(r"(https?|ftp|file)://[-A-Za-z0-9+&@#/%?=~_|!:,.;]+[-A-Za-z0-9+&@#/%=~_|]","",text)
    
    # delete all the non-English characters  
    res = re.compile("[^a-z^A-Z^0-9]")
    text = res.sub(r" ", text)
    
    # tokenize
    text = nltk.word_tokenize(text)
    
    # word segmentation
    exception = ["makeupforever", "anatasiabeverlyhills", "nyx"]
    tweet_processed_list = [seg.segment(word) for word in text if word not in exception]
    
    # delete stopwords
    filtered = [w for w in text if w not in stopwords.words('english')]
    
    # stemmisation (remove for now)
    # ps = PorterStemmer()
    # filtered = [ps.stem(w) for w in filtered]
    
    # lemmanisation
    lemmatizer = nltk.stem.WordNetLemmatizer()
    filtered = [lemmatizer.lemmatize(w) for w in filtered]
    
    # only use NN to keep nouns
    refiltered = nltk.pos_tag(filtered)
    filtered = [w for w, pos in refiltered if pos.startswith('NN')]

    return " ".join(filtered)
```

```python
WETNWILD2019_list = []
for i in range(0, len(WETNWILD2019)):
    WETNWILD2019_list.append(textPreprocessing(WETNWILD2019[i]))
    
WETNWILD2020_list = []
for i in range(0, len(WETNWILD2020)):
    WETNWILD2020_list.append(textPreprocessing(WETNWILD2020[i]))


WETNWILD2021_list = []
for i in range(0, len(WETNWILD2021)):
    WETNWILD2021_list.append(textPreprocessing(WETNWILD2021[i]))
```

```python
with open('newsdata.txt','a+', encoding='utf-8') as f:
    for tweet in WETNWILD2019_list:
        f.write((str(tweet)+'\n'))
    for tweet in WETNWILD2020_list:
        f.write((str(tweet)+'\n'))
    for tweet in WETNWILD2021_list:
        f.write((str(tweet)+'\n'))
```

## Dynamic Topic Model

```python
#logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
stoplist = set()
dictionary=corpora.Dictionary(line.lower().split() for line in open('newsdata.txt'))
stop_ids=[
    dictionary.token2id[stopword]
    for stopword in stoplist
    if stopword in dictionary.token2id]

once_ids=[tokenid for tokenid, docfreq in iteritems(dictionary.dfs) if docfreq ==1]
dictionary.filter_tokens(stop_ids+once_ids)
```

```python
dictionary.compactify()
dictionary.save('news_dictionary')

class MyCorpus(object):
    def __iter__(self):
        for line in open('newsdata.txt'):
            yield dictionary.doc2bow(line.lower().split())
            
corpus_memory_friendly = MyCorpus()
corpus = [vector for vector in corpus_memory_friendly]
corpora.BleiCorpus.serialize('news_corpus',corpus)
```

```python
try:
    dictionary=Dictionary.load('news_dictionary')
except FileNotFoundError as e:
    raise ValueError("Didn't find the dictionary")
corpus=bleicorpus.BleiCorpus('news_corpus')
time_slice=[5200, 2566, 1002]
num_topics=5
ldaseq=ldaseqmodel.LdaSeqModel(corpus=corpus, id2word=dictionary, time_slice=time_slice, num_topics=num_topics)
```

## Visulisation

```python
LEN = [5200,2566,1002]
```

```python
topictime = []
for i in range(len(LEN)):
    b = []
    if i == 0:
        a = 0
    else:
        a = sum(LEN[k] for k in range(i))  
    for j in range(LEN[i]):
        doc = ldaseq.doc_topics(a+j)
        rank = np.argsort(doc)
        b.extend(rank[-1:])
    topictime.append(b)
```

```python
topicdis = [[0 for i in range(num_topics)] for j in range(len(topictime))]
for i in range(len(topictime)):
    c = collections.Counter(topictime[i])
    c = dict(c)
    for j in range(len(c)):
        topicdis[i][j] = c[j]/LEN[i]
```

### Fig.3. perplexity score for different number of topics 

```python
WETWILD_doc = []
WETWILD_doc.extend(WETNWILD2019_list)
WETWILD_doc.extend(WETNWILD2020_list)
WETWILD_doc.extend(WETNWILD2021_list)
print(len(WETWILD_doc))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
tf_vectorizer = CountVectorizer(max_df=0.95, 
                                min_df=2,          # minimum reqd occurences of a word
                                max_features=2000,  # max number of uniq words
                                stop_words='english')
tf = tf_vectorizer.fit_transform(WETWILD_doc)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

n_topics = range(5, 30, 5)
perplexityLst = [1.0]*len(n_topics)

#训练LDA并打印训练时间
lda_models = []
for idx, n_topic in enumerate(n_topics):
    lda = LatentDirichletAllocation(n_components=n_topic,
                                    max_iter=20,
                                    learning_method='batch',
                                    evaluate_every=200,
#                                    perp_tol=0.1, #default                                       
#                                    doc_topic_prior=1/n_topic, #default
#                                    topic_word_prior=1/n_topic, #default
                                    verbose=0)

    lda.fit(tf)
    perplexityLst[idx] = lda.perplexity(tf)
    lda_models.append(lda)
    print ("# of Topic: %d, " % n_topics[idx],)
    print ("Perplexity Score %0.3f" % perplexityLst[idx])

#打印最佳模型
best_index = perplexityLst.index(min(perplexityLst))
best_n_topic = n_topics[best_index]
best_model = lda_models[best_index]
print ("Best # of Topic: ", best_n_topic)
```

```python
#绘制不同主题数perplexity的不同
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(n_topics, perplexityLst)
ax.set_xlabel("# of topics")
ax.set_ylabel("Approximate Perplexity")
plt.grid(True)
#plt.savefig(os.path.join('lda_result', 'perplexityTrend'+'.png'))
plt.show()
```
