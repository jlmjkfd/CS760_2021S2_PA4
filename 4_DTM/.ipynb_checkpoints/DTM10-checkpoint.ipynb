```python
# setting up our imports
import logging
from six import iteritems
from gensim import corpora
from gensim.models import ldaseqmodel
from gensim.corpora import Dictionary, bleicorpus
from gensim.matutils import hellinger
```

```python
#logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
stoplist = set()
dictionary=corpora.Dictionary(line.lower().split() for line in open('newsdata_nobrand.txt'))
stop_ids=[
    dictionary.token2id[stopword]
    for stopword in stoplist
    if stopword in dictionary.token2id]

once_ids=[tokenid for tokenid, docfreq in iteritems(dictionary.dfs) if docfreq ==1]
dictionary.filter_tokens(stop_ids+once_ids)
```

```python
dictionary.compactify()
dictionary.save('news_dictionary')

class MyCorpus(object):
    def __iter__(self):
        for line in open('newsdata_nobrand.txt'):
            yield dictionary.doc2bow(line.lower().split())
            
corpus_memory_friendly = MyCorpus()
corpus = [vector for vector in corpus_memory_friendly]
corpora.BleiCorpus.serialize('news_corpus',corpus)
```

```python
# 调seed！
try:
    dictionary=Dictionary.load('news_dictionary')
except FileNotFoundError as e:
    raise ValueError("Didn't find the dictionary")
corpus=bleicorpus.BleiCorpus('news_corpus')
time_slice=[47244,44777,39429,41978,37574,34968,43394,57018,18952,20947]
num_topics=20
ldaseq=ldaseqmodel.LdaSeqModel(corpus=corpus, id2word=dictionary, time_slice=time_slice, alphas = 50/num_topics, chain_variance = 0.01, num_topics=num_topics, random_state = 10)
```

```python
corpusTopic=ldaseq.print_topics(time=0)
print(corpusTopic)
topicEvolution = ldaseq.print_topic_times(topic=0)
print(topicEvolution)
doc = ldaseq.doc_topics(0)
print(doc)
```

```python
corpusTopic=ldaseq.print_topics(time=1)
print(corpusTopic)
```

```python
topicEvolution = ldaseq.print_topic_times(topic=1)
print(topicEvolution)
```

```python
LEN = time_slice
```

```python
topictime = []
for i in range(len(LEN)):
    b = []
    if i == 0:
        a = 0
    else:
        a = sum(LEN[k] for k in range(i))  
    for j in range(LEN[i]):
        doc = ldaseq.doc_topics(a+j)
        rank = np.argsort(doc)
        b.extend(rank[-1:])
    topictime.append(b)
```

```python
topicdis = [[0 for i in range(num_topics)] for j in range(len(topictime))]
for i in range(len(topictime)):
    c = collections.Counter(topictime[i])
    c = dict(c)
    for j in range(len(c)):
        topicdis[i][j] = c[j]/LEN[i]
```

# Visualization

## Fig.1.a.topic proportion over time (bar chart)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize=(32,16))
year = ['19Q1','19Q2','19Q3','19Q4','20Q1','20Q2','20Q3','20Q4','21Q1','21Q2']
col = ['#63b2ee','#76da91','#f8cb7f','#f89588','#7cd6cf','#9192ab','#7898e1','#efa666','#eddd86','#9987ce',
         '#95a2ff','#fa8080','#ffc076','#fae768','#87e885','#3cb9fc','#73abf5','#cb9bff','#90ed7d','#f7a35c']
topicdis1 = DataFrame(topicdis)
topic2 = topicdis1.T
topic2 = np.array(topic2)
for i in range(20):
    data = topic2[i]
    if i == 0:
        plt.bar(year,data)
    else:
        bot = sum(topic2[k] for k in range(i)) 
        plt.bar(year,data,color=col[i],bottom = bot)

# x and y limits
#plt.xlim(-0.6, 2.5)
#plt.ylim(-0.0, 1.5)
# remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
#grid
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.7)
# title and legend
legend_label = ['Topic1', 'Topic2', 'Topic3', 'Topic4','Topic5','Topic6','Topic7','Topic8','Topic9','Topic10','Topic11'
               ,'Topic12','Topic13','Topic14','Topic15','Topic16','Topic17','Topic18','Topic19','Topic20']
plt.legend(legend_label, ncol = 20, bbox_to_anchor=([1, 1, 0, 0]), frameon = False)
plt.title('Topic proportions over 2019 - 2021\n', loc='left', fontsize = 20)
plt.show()
```

## Fig.2. topic key words over time

```python
topicEvolution0 = ldaseq.print_topic_times(topic=0)
```

```python
fig, axes = plt.subplots(2,5, figsize=(30, 15), sharex=True)
axes = axes.flatten()
year = ['19Q1','19Q2','19Q3','19Q4','20Q1','20Q2','20Q3','20Q4','21Q1','21Q2']
title = 'Topic0'
for i in range(len(topicEvolution0)):
    value = [item[1] for item in topicEvolution0[i]]
    index = [item[0] for item in topicEvolution0[i]]
    
    ax = axes[i]
    ax.barh(index,value,height = 0.7)
    ax.set_title(year[i],
                     fontdict={'fontsize': 30})
    ax.invert_yaxis()
    ax.tick_params(axis='both', which='major', labelsize=20)
    for k in 'top right left'.split():
        ax.spines[k].set_visible(False)
    fig.suptitle(title, fontsize=40)

plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
plt.show()
```
